# Load LAKH Dataset

```python
import tarfile

LAKH_MIDI_METADATA_COMPRESSED_PATH = 'lakh-midi-dataset_matched_metadata_h5.tar.gz'
LAKH_MIDI_METADATA_PATH = 'lmd_matched_h5'


with tarfile.open(LAKH_MIDI_METADATA_COMPRESSED_PATH, "r:gz") as tar:
    tar.extractall()
```

```python
! chmod -R 777 lmd_matched_h5/
```

```python
import os

MAX_PATHS_TO_PRINT = 15


printed_paths = 0
for root, dirs, files in os.walk(LAKH_MIDI_METADATA_PATH):
  for filename in files:
    if filename.endswith(".h5"):
      print(os.path.join(root, filename))
      printed_paths+=1
      if printed_paths >= MAX_PATHS_TO_PRINT:
        break
  if printed_paths >= MAX_PATHS_TO_PRINT:
    break
```

```python
import tables as tb


lakh_metadata_dict = {'filename':[]}
for root, dirs, files in os.walk(LAKH_MIDI_METADATA_PATH):
  for filename in files:
    if filename.endswith(".h5"):
      filepath = os.path.join(root, filename)
      with tb.open_file(filepath, mode="r") as h5file:
        # print(h5file.title)
        table = h5file.root.metadata.songs
        table_columns = table.colnames
        for row in table:
          lakh_metadata_dict['filename'].append(filename)
          for columnn in table_columns:
            if columnn not in lakh_metadata_dict:
              lakh_metadata_dict[columnn] = []
            value = row[columnn]
            if type(value) not in (int,float,):
              value=value.decode('utf-8')
            lakh_metadata_dict[columnn].append(value)
```

```python
import pandas as pd

lakh_metadata_df = pd.DataFrame.from_dict(lakh_metadata_dict)
lakh_metadata_df.head()
```

```python
import numpy as np

for col in list(lakh_metadata_df.columns):

  lakh_metadata_df[col].replace('', np.nan, inplace=True)
```

```python
lakh_metadata_df.dropna(how='all', axis=1, inplace=True)
lakh_metadata_df.head()
```

```python
print('Columns')
for c in list(lakh_metadata_df.columns):
  print(f'\t{c}')

print(f'Size: {len(lakh_metadata_df)}')
```

## Organize Midi Files

```python
filenames=set(merged_metadata_df['filename'])
print(len(filenames))
print(len(set(filenames)))
list(filenames)[:10]
```

```python
midi_filepaths = {}
for root, dirs, _ in os.walk(LAKH_MIDI_PATH):
  for dir in dirs:
    if dir in filenames:
      for sub_root, _, files in os.walk(os.path.join(root, dir)):
        for filename in files:
          if filename.endswith(".mid") or filename.endswith(".midi"):
            midi_filepaths[dir] = os.path.join(root, dir, filename)
print(len(midi_filepaths))
```

```python
list(midi_filepaths.values())[:3]
```

```python
names_and_genre_df = sampled_df[['filename', 'genre']]
names_and_genre_df
```

```python
! rm -rf midi_songs
```

```python
MIDI_SONGS_OUTPUT_FOLDER = 'midi_songs'

filename_to_midi = {}
midi_locations_per_genre = {}
for filename, genre in names_and_genre_df.itertuples(index=False,name=None):
  if genre not in midi_locations_per_genre:
    midi_locations_per_genre[genre] = []
  midi_locations_per_genre[genre].append(midi_filepaths[filename])
  filename_to_midi[filename]=os.path.basename(midi_filepaths[filename])
```

```python
for k,v in midi_locations_per_genre.items():
  print(f'{k}: {len(set(v))}')
```

```python
for genre in midi_locations_per_genre.keys():
  os.makedirs(os.path.join(MIDI_SONGS_OUTPUT_FOLDER,genre), exist_ok=True)
```

```python
import shutil

copied_songs = []
for genre, songs in midi_locations_per_genre.items():
  for song in songs:
    filename = os.path.basename(song)
    copied_songs.append(filename)
    dst =  os.path.join(MIDI_SONGS_OUTPUT_FOLDER, genre, filename)
    shutil.copyfile(song, dst)
    # print(song,  '--->',dst)
print(len(copied_songs))
print(len(set(copied_songs)))
```

```python
duplicated_midis = set([x for x in copied_songs if copied_songs.count(x) > 1])
```
